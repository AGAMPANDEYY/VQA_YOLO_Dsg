```python
from open_flamingo import create_model_and_transforms

model,image_processor,tokenizer=create_model_and_transforms(
    clip_vision_encoder_path="ViT-L-14",
    clip_vision_encoder_pretrained="openai",
    lang_encoder_path="anas-awadalla/mpt-1b-redpajama-200b",
    tokenizer_path="anas-awadalla/mpt-1b-redpajama-200b",
    cross_attn_every_n_layers=1,
)
```

```python
from huggingface_hub import hf_hub_download
import torch

checkpoint_path = hf_hub_download("openflamingo/OpenFlamingo-4B-vitl-rpj3b-langinstruct", "checkpoint.pt")
model.load_state_dict(torch.load(checkpoint_path), strict=False)
```

```python
from PIL import Image
import requests
import torch


#load image for QA

img1=Image.open("/content/frame0-00-10.01.jpg")
```

Pre process the data since
etails: For OpenFlamingo, we expect the image to be a torch tensor of shape
 batch_size x num_media x num_frames x channels x height x width.
 In this case batch_size = 1, num_media = 3, num_frames = 1,
 channels = 3, height = 224, width = 224.

```python
vision_x=image_processor(img1).unsqueeze(0)
vision_x=vision_x.unsqueeze(1).unsqueeze(0)
```

```python
vision_x
```

```python
tokenizer.padding_side='left'
lang_x=tokenizer(text="The number of cars here are",return_tensors='pt')
```

```python
generated_text=model.generate(
    vision_x=vision_x,
    lang_x=lang_x["input_ids"],
    attention_mask=lang_x['attention_mask'],
    max_new_tokens=30,
    num_beams=3,
)
```

```python
generated_text
```

```python
output_response=tokenizer.decode(generated_text[0])
```

```python
print("The generated text by Open-flamingo is")
print("--------------------------------------")
print(output_response)
```

```python
output=tokenizer.decode(generated_text[0])
```
